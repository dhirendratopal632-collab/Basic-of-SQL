Q1. Create a New Database and Table for Employees


```python
CREATE DATABASE company_db;
USE company_db;

CREATE TABLE employees (
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    department VARCHAR(50),
    salary INT,
    hire_date DATE
);
```

Q3. Display All Employee Records Sorted by Salary (Lowest to Highest)

```python
SELECT *
FROM employees
ORDER BY salary ASC;
```

Q4. Show Employees Sorted by Department (A–Z) and Salary (High → Low)

```python
SELECT *
FROM employees
ORDER BY department ASC, salary DESC;
```

Q5. List All Employees in the IT Department, Ordered by Hire Date (Newest First)

```python
SELECT *
FROM employees
WHERE department = 'IT'
ORDER BY hire_date DESC;
```

Q9. Difference Between Primary Key and Foreign Key

Primary Key

Uniquely identifies each record in a table

Cannot contain NULL values

One primary key per table

Foreign Key

Refers to the primary key of another table

Can contain duplicate values

Used to maintain relationship between tables

Q10. What Are Constraints in SQL and Why Are They Used?

Constraints are rules applied to table columns to ensure data accuracy and integrity.

Common Constraints:

PRIMARY KEY – uniquely identifies a record

NOT NULL – prevents NULL values

UNIQUE – ensures unique values

CHECK – applies a condition

FOREIGN KEY – maintains table relationships

Why used:
To prevent invalid data, enforce business rules, and maintain database consistency.
